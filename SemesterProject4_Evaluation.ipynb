## **Anderson-Darling - Normality Test**

```python
print(stats.anderson(dataB.time))

#Example of Anderson-Darling Test - Group A 
print("Group A - Time ")
result = stats.anderson(dataA.time)

print('stat=%.3f' % (result.statistic))
for i in range(len(result.critical_values)):
 sl, cv = result.significance_level[i], result.critical_values[i]
 if result.statistic < cv:
  print('Data follows Normal at the %.1f%% level' % (sl))
 else:
  print('Data does not follows Normal at the %.1f%% level' % (sl))

print()
#Example of Anderson-Darling Test - Group B
print("Group B - Time")
result2 = stats.anderson(dataB.time)
print('stat=%.3f' % (result2.statistic))
for ii in range(len(result2.critical_values)):
 sl2, cv2 = result2.significance_level[ii], result2.critical_values[ii]
 if result2.statistic < cv2:
  print('Data follows Normal at the %.1f%% level' % (sl2))
 else:
  print('Data does not follows Normal at the %.1f%% level' % (sl2))

print()
#Example of Anderson-Darling Test - Group B
print("Group A - Coin")
result3 = stats.anderson(dataA.coin)
print('stat=%.3f' % (result3.statistic))
for iii in range(len(result3.critical_values)):
 sl23, cv23 = result3.significance_level[iii], result3.critical_values[iii]
 if result3.statistic < cv23:
  print('Data follows Normal at the %.1f%% level' % (sl23))
 else:
  print('Data does not follows Normal at the %.1f%% level' % (sl23))

print()
#Example of Anderson-Darling Test - Group B
print("Group B - Coin")
result4 = stats.anderson(dataB.coin)
print('stat=%.3f' % (result4.statistic))
for iiii in range(len(result4.critical_values)):
 sl234, cv234 = result4.significance_level[iiii], result4.critical_values[iiii]
 if result4.statistic < cv234:
  print('Data follows Normal at the %.1f%% level' % (sl234))
 else:
  print('Data does not follows Normal at the %.1f%% level' % (sl234))
```

An Anderson-Darling Test is a goodness of fit test that measures how well your data fit a specified distribution. This test is most commonly used to determine whether or not your data follow a normal distribution. This test does not need a huge amount of samples in order to work. 

The null hypothesis is that the data are normally distributed; the alternative hypothesis is that the data are non-normal.

For example:
The test statistic for time group A, is 0.275. We can compare this value to each critical value that corresponds to each significance level to see if the test results are significant. For example:

The critical value for α = 0.01 is 0.961 Because the test statistic (0.275) is not greater than this critical value, the results are not significant at a significance level of 0.01.
The critical value for α = 0.025 is 0.808. Because the test statistic (0.275) is not greater than this critical value, the results are not significant at a significance level of 0.025.

We can see that the test results are not significant at any significance level, which means we would not reject the null hypothesis of the test. Thus, we don’t have sufficient evidence to say that the sample data is not normally distributed.

However, the coins in group A does clearly not follow a normal distribution since the test statistic is significant at all significant levels. Time group A could be normally distributed since its not significant at any significant levels. 

The other groups lay around 0.600, and they are only significant at the upper most significant levels 15% and 10%. 

It is therefore not possible to conlcude that the data is normally distributed, but some of it is and some of it is not.

## **Shapiro Wilks - Normality Test**

The null-hypothesis of this test is that the population is normally distributed. Thus, if the p value is less than the chosen alpha level, then the null hypothesis is rejected and there is evidence that the data tested are not normally distributed. On the other hand, if the p value is greater than the chosen alpha level, then the null hypothesis (that the data came from a normally distributed population) can not be rejected (e.g., for an alpha level of .05, a data set with a p value of less than .05 rejects the null hypothesis that the data are from a normally distributed population).

```python
print("Time - Group A")
print(stats.shapiro(dataA.time))
print()
print("Time - Group B")
print(stats.shapiro(dataB.time))
print()
print("Coin - Group A")
print(stats.shapiro(dataA.coin))
print()
print("Coin - Coin B")
print(stats.shapiro(dataB.coin))
print()
```

As we can see here. The Shaprio Wilks test confirms what the Anderson test also confirmed. Time Group A cannot reject the null-hypothesis and therefore it can be from a normal distribution. 

However, time group B and all coin groups are not from normal distributions, as the p-value is less then the chosen alpha level of 0.05

## ***Correlation Tests***

## Spearman’s Rank Correlation - Correleation between two variables (non parametric) 

The Spearman rank-order correlation coefficient is a nonparametric measure of the monotonicity of the relationship between two datasets. Unlike the Pearson correlation, the Spearman correlation does not assume that both datasets are normally distributed. Like other correlation coefficients, this one varies between -1 and +1 with 0 implying no correlation. Correlations of -1 or +1 imply an exact monotonic relationship. Positive correlations imply that as x increases, so does y. Negative correlations imply that as x increases, y decreases.

```python
statn, pn = stats.spearmanr(dataA.time, dataB.time)
print('stat=%.3f, p=%.3f' % (statn, pn))
if pn > 0.05:
	print('Probably independent')
else:
	print('Probably dependent')
 
```

## **Kruskal-Wallis H Test - Distribution Test**


The Kruskal-Wallis test is a nonparametric version of the one-way analysis of variance test or ANOVA for short. It is named for the developers of the method, William Kruskal and Wilson Wallis. This test can be used to determine whether more than two independent samples have a different distribution. It can be thought of as the generalization of the Mann-Whitney U test.

The default assumption or the null hypothesis is that all data samples were drawn from the same distribution  or identical population with the same or identical median.

Specifically, that the population medians of all groups are equal. A rejection of the null hypothesis indicates that there is enough evidence to suggest that one or more samples dominate another sample, but the test does not indicate which samples or by how much. This test does not need a huge amount of samples in order to work. Each data sample must be independent, have 5 or more observations, and the data samples can differ in size.

```python
# Kruskal-Wallis H-test - Time

# compare samples
stat2, p2 = stats.kruskal(dataA.time, dataB.time)
print('Statistics=%.3f, p=%.3f' % (stat2, p2))
# interpret
alpha2 = 0.05
if p2 > alpha2:
	print('Same distributions (fail to reject H0)')
else:
	print('Different distributions (reject H0)')
```

The p-value is interpreted, correctly rejecting the null hypothesis that all samples have the same distribution. Meaning that the time data comes from the same distribution. 

```python
# Kruskal-Wallis H-test - Coins

# compare samples
stat3, p3 = stats.kruskal(dataA.coin, dataB.coin)
print('Statistics=%.3f, p=%.3f' % (stat3, p3))
# interpret
alpha3 = 0.05
if p3 > alpha3:
	print('Same distributions (fail to reject H0)')
else:
	print('Different distributions (reject H0)')
```

## **Mann Whitney Test - Median Difference Testing**

The test compares two populations. The null hypothesis for the test is that the probability is 50% that a randomly drawn member of the first population will exceed a member of the second population or,  Another option for the null hypothesis is that the two samples come from the same population (i.e. that they both have the same median)

H0: η1 = η2          
H1: η1 < η2           
H1: η1 > η2	      
H1: η1 ≠ η2	


```python
print(stats.mannwhitneyu(dataA.time,dataB.time))
print(stats.mannwhitneyu(dataA.coin,dataB.coin))
```

The sample size is not nearly big enough to complete a proper Mann Whitney. But since the test for both coins and time is not significant, we cannot reject the null hypothesis both samples have the same median. 

**Results explanation:** 

If the p-value is less than or equal to the significance level, the decision is to reject the null hypothesis. You can conclude that the difference between the population medians is statistically significant.

If the p-value is greater than the significance level, the decision is to fail to reject the null hypothesis. We do not have enough evidence to conclude that the difference between the population medians is statistically significantly.

Thus we cannot reject the null hypothesis, that the two samples come from the same population/have the same median and therefore they might be a difference in the medians of the two samples, which works in favor of our test.
